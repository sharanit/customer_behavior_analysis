# AEROFIT TREADMILL CUSTOMER ANALYSIS - SUMMARY
## Business Case Study Report

---

## EXECUTIVE SUMMARY

This comprehensive analysis examines customer purchasing patterns for Aerofit's three treadmill models (KP281, KP481, and KP781) to develop targeted customer profiles and provide actionable business recommendations. The study analyzes demographic, socioeconomic, and behavioral characteristics of 180 customers over a three-month period.

**Key Findings:**

1. KP281 (Entry-level, $1,500): Most popular, targeting younger, budget-conscious customers

2. KP481 (Mid-level, $1,750): Balanced market segment with moderate fitness enthusiasts

3. KP781 (Premium, $2,500): High-income, fitness-focused customers with intensive usage patterns

---

## 1. PROBLEM STATEMENT

**Business Objective:** Identify the characteristics of the target audience for each treadmill product to provide better recommendations to new customers and optimize marketing strategies.

**Research Questions:**
1. What are the demographic profiles of customers for each product?
2. How do income, age, and fitness levels influence product choice?
3. What is the probability distribution across different customer segments?
4. What actionable insights can drive sales and customer satisfaction?

## 6. ACTIONABLE RECOMMENDATIONS

### 6.1 MARKETING & POSITIONING

**Recommendation 1: Targeted Gender Marketing for KP781**
- **Action:** Create female-focused marketing campaign for premium model
- **Rationale:** Only 18% female buyers in premium segment vs 48% in entry level
- **Tactics:**
  - Feature female athletes and influencers
  - Highlight features appealing to women (safety, convenience, design)
  - Partner with women's fitness communities
- **Expected Impact:** Increase female KP781 buyers by 50% (from 18% to 27%)

**Recommendation 2: Income-Based Segmentation**
- **Action:** Develop three distinct marketing campaigns based on income brackets
  - Under $50K → KP281 focus
  - $50K-$75K → KP481 focus
  - Over $75K → KP781 focus
- **Rationale:** Income is strongest predictor (correlation 0.78)
- **Tactics:**
  - Digital advertising with income targeting
  - LinkedIn ads for high-income professionals
  - Student/young professional discounts for KP281
- **Expected Impact:** 15-20% improvement in conversion rates

**Recommendation 3: Partner-Focused Promotions**
- **Action:** "Couples Fitness Package" with household discounts
- **Rationale:** 59% buyers are partnered; partnered buyers prefer premium models
- **Tactics:**
  - "Buy together, save together" promotions
  - Family fitness marketing
  - Referral programs for couples
- **Expected Impact:** Increase partnered customer acquisitions by 25%

### 6.2 PRODUCT STRATEGY

**Recommendation 4: Introduce KP500 - Mid-Premium Model**
- **Action:** Create new model at $2,000 price point
- **Rationale:** Large gap between KP481 ($1,750) and KP781 ($2,500)
- **Target:** Customers earning $65K-$80K with fitness level 3.5-4.0
- **Expected Impact:** Capture 10-15% of hesitant premium buyers

**Recommendation 5: Feature Differentiation**
- **Action:** Emphasize specific features for each segment
  - KP281: Affordability, space-saving, basic tracking
  - KP481: Balance of features and value, training programs
  - KP781: Advanced metrics, durability, professional-grade
- **Rationale:** Clear differentiation improves decision-making
- **Expected Impact:** Reduce choice confusion, improve satisfaction

### 6.3 SALES & CUSTOMER EXPERIENCE

**Recommendation 6: Predictive Recommendation System**
- **Action:** Implement customer profiling tool for sales staff
- **Input Variables:** Age, income, fitness level, usage goals
- **Output:** Recommended product with probability scores
- **Example:** 32-year-old, $70K income, fitness level 3.5, 3x/week usage → 75% probability KP481 fit
- **Expected Impact:** Improve recommendation accuracy by 40%

**Recommendation 7: Trade-Up Program**
- **Action:** Launch "Fitness Journey Upgrade" program
- **Rationale:** Entry-level buyers may upgrade as fitness improves
- **Tactics:**
  - Trade-in value on old treadmill (up to $500)
  - Loyalty discount for existing customers (15% off)
  - Track customer usage data to identify upgrade candidates
- **Expected Impact:** Increase upgrade rate by 30%

**Recommendation 8: Financing Options**
- **Action:** Offer 0% financing for 12-24 months
- **Target:** KP481 and KP781 models
- **Rationale:** Reduce price barrier for qualified buyers
- **Qualification:** Income >$50K, credit check
- **Expected Impact:** Increase mid/premium sales by 20-25%

### 6.4 CONTENT & EDUCATION

**Recommendation 9: Fitness Level-Based Content Marketing**
- **Action:** Create content series for each fitness level
  - Level 1-2: "Starting Your Fitness Journey" (KP281 focus)
  - Level 3: "Taking It to the Next Level" (KP481 focus)
  - Level 4-5: "Peak Performance Training" (KP781 focus)
- **Channels:** Blog, YouTube, email campaigns
- **Expected Impact:** Improve brand engagement and informed purchases

**Recommendation 10: Virtual Try-Before-Buy**
- **Action:** Implement AR/VR showroom and 30-day trial program
- **Rationale:** High involvement purchase benefits from trial
- **Tactics:**
  - Virtual reality showroom experience
  - 30-day money-back guarantee
  - In-home trial program (refundable deposit)
- **Expected Impact:** Reduce purchase hesitation, improve conversion by 15%

### 6.5 RETENTION & LOYALTY

**Recommendation 11: Usage-Based Rewards Program**
- **Action:** "Aerofit Miles Club" - earn points based on usage
- **Rationale:** High correlation between usage and satisfaction

**OUTLIER DETECTION**

```python
numerical_cols = ['Age', 'Education', 'Usage', 'Income', 'Fitness', 'Miles']

fig, axes = plt.subplots(2, 3, figsize=(18, 9))
axes = axes.ravel()

for idx, col in enumerate(numerical_cols):

    ax = axes[idx]

    # Draw Boxplot
    bp = ax.boxplot(df[col], vert=True, patch_artist=True)
    ax.set_title(f'{col} - Boxplot', fontsize=12, fontweight='bold')
    ax.set_ylabel(col)

    # Color the box
    for patch in bp['boxes']:
        patch.set_facecolor('lightblue')

    # Calculate Outliers using IQR method
    Q1 = df[col].quantile(0.25)
    Q3 = df[col].quantile(0.75)
    IQR = Q3 - Q1
    lower_bound = Q1 - 1.5 * IQR
    upper_bound = Q3 + 1.5 * IQR
    outliers = df[(df[col] < lower_bound) | (df[col] > upper_bound)][col]

    # PRINT STATS INSIDE THE SUBPLOT
    text = (
        f"Q1: {Q1:.2f}\n"
        f"Q3: {Q3:.2f}\n"
        f"IQR: {IQR:.2f}\n"
        f"Lower Bound: {lower_bound:.2f}\n"
        f"Upper Bound: {upper_bound:.2f}\n"
        f"Outliers: {len(outliers)}"
    )

    ax.annotate(
        text,
        xy=(1.15, 0.85),
        xycoords='axes fraction',
        fontsize=10,
        va='center',
        bbox=dict(boxstyle="round,pad=0.3", fc="whitesmoke", ec="gray")
    )

plt.tight_layout()
plt.savefig('outlier_detection_side_by_side.png', dpi=300, bbox_inches='tight')
plt.show()
```
